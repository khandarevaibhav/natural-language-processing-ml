```python
import pandas as pd
```

```python
ds = pd.read_csv("Restaurant_Reviews.tsv", delimiter = "\t")
```

```python
ds 
```

```python
reviews = ds['Review']
```

```python
reviews
```

```python
import re
```

```python
import nltk
```

```python
corpus = []
```

```python
for i in reviews:
    review = re.sub('[^a-zA-Z]', ' ', i)
    review = review.lower()
    corpus.append(review)
    
    
    
```

```python
corpus
```

```python
from nltk.stem import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
ps.stem("getting")
```

```python
for i in reviews:
    review = re.sub('[^a-zA-Z]', ' ', i)
    review = review.lower()
    words = review.split()
    stem_word = []
    for word in words:
        
        x = ps.stem(word)
        stem_word.append(x)
    review = ' '.join(stem_word)    
    corpus.append(review)
    
```

```python
corpus
```

```python
nltk.download("stopwords")
```

```python
from nltk.corpus import stopwords
```

```python
all_stopwords = stopwords.words("english")
```

```python
all_stopwords
```

```python
all_stopwords.remove("not")
```

```python
all_stopwords
```

```python
for i in reviews:
    review = re.sub('[^a-zA-Z]', ' ', i)
    review = review.lower()
    words = review.split()
    stem_word = []
    for word in words:
        if word not in all_stopwords:
            x = ps.stem(word)
            stem_word.append(x)
    review = ' '.join(stem_word)    
    corpus.append(review)
    
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#bag of 
bog = CountVectorizer(max_features=1550)
```

```python
X = bog.fit_transform(corpus)
```

```python
#sentence to numpy array
X = X.toarray()
```

```python
X.shape
```

```python
X[2]
```

```python
y = ds['Liked'].values
```

```python
y.shape
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB
```

```python
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

```python
y_test
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_test, y_pred)
```
