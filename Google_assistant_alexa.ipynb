```python
import pyttsx3

engine = pyttsx3.init()
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
speak("Hello Vaibhav") 
```

```python
#to change assitent voice 
engine = pyttsx3.init()
voice = engine.getProperty('voices')
engine.setProperty('voice', voice[1].id) #0 for male and 1 for femal voice
newVoiceRate = 150 #default is this is for controlling voice speed
engine.setProperty('rate', newVoiceRate)
def speak(audio):
    engine.say(audio)
    engine.runAndWait()
speak("Hello Vaibhav")
```

```python
import datetime
def time():
    Time = datetime.datetime.now().strftime("%I:%M:%S")
    speak(Time)
time()    
```

```python
def date():
    year = int(datetime.datetime.now().year)
    month = int(datetime.datetime.now().month)
    date = int(datetime.datetime.now().day)
    speak("Todays Date is ...")
    speak(date)
    speak(month)
    speak(year)
    
date()    
```

```python
def wishme():
    speak("Welcome Back Vaibhav!")
    #time() # because date and time we are calling into if else in main
   # date()
   
    hour = datetime.datetime.now().hour
    
    if hour >= 6 and hour < 12:
        speak("Good Morning Vaibhav")
    elif hour >=12 and hour < 18:
        speak("Good Afternoon Vaibhav")
    elif hour >=18 and hour <= 24:
        speak("Good Evening Vaibhav")
    else:
        speak("Good Night Vaibhav")
    speak("Today at your service. How I Can Help You ?")    
wishme()       
```

```python
import speech_recognition as sr
r = sr.Recognizer()
def takeCommand():
    with sr.Microphone() as source:
        print("Start Saying...")
        audio = r.listen(source)
        print("Speech is Done")
    #query=r.recognize_google(audio, 'en=US')    
    
    try:
        print("Recognizing.....")
        query = r.recognize_google(audio, language="en")
        print(query)
    except Exception as e:
        speak(e)
        speak("Say that again please")
        
        return "None"
    
    return query
takeCommand()    
```

```python
if __name__ == "__main__":
    wishme()
    while True:
        input_voice = takeCommand().lower() # all test convert into lower case because it not create problem in if-else
        print(input_voice)
        
        if "time" in input_voice:
            time()
        elif "date" in input_voice:
            date()
        elif "offline" in input_voice:
            quit()
        
```
